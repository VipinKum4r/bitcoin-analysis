2. Look at some information about the data and the columns

```python
import pandas as pd #imported pandas so that we can use the pandas library
```

```python
df=pd.read_csv("bitcoin_data.csv") #importing the csv dataset
```

```python
df #printing the first 5 and the last 5 rows of the dataset
```

```python
len(df.columns) #number of columns
```

```python
len(df) #number of rows
```

```python
df.info() #overall information
```

```python
df.describe() #gives mean,mode,median and other basic statistical values
```

```python
numerics = ['int16','int32','int64','float16','float32','float64'] #tells how many numeric columns do we have
numeric_df=df.select_dtypes(include=numerics)
len(numeric_df.columns)
```

```python
missing_percentages=df.isna().sum().sort_values(ascending=False)/len(df) #tells how many missing values do we have
missing_percentages
```

```python
missing_percentages[missing_percentages!=0].plot(kind="barh") #remove the columns with no values in missing percentages
```
